```python
shark_data = pd.read_excel("GSAF5.xls")
```

- Since the dataset is not a csv file, and is an excel file, I used read_excel instead.

```python
shark_data = shark_data.dropna(axis=1, how='all')
```

- An observation is a row in the dataset that provides information. In this dataset, an observation represents one shark attack. If a row/observation is missing or inconsistent it is more efficient to get rid of it to clean the data. 

```python
shark_data['Year'] = pd.to_numeric(shark_data['Year'], errors='coerce')
shark_data = shark_data[(shark_data['Year'] >= 1940) & (shark_data['Year'] <= 2025)]
print("Year range:", shark_data['Year'].min(), "-", shark_data['Year'].max())
shark_data = shark_data[shark_data['Year'] >= 1940]
```

```python
attacks_per_year = shark_data.groupby('Year').size()
plt.figure(figsize=(12,5))
attacks_per_year.plot()
plt.title("Shark Attacks per Year Since 1940")
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.show()
```

- This plot shows that the shark attacks since 1940 experienced a decrease from 1960-1980 but then started increasing before decreasing again starting 2019. 

```python
shark_data['Age'] = pd.to_numeric(shark_data['Age'], errors='coerce')
ages = shark_data['Age'].dropna()
plt.figure(figsize=(10,5))
plt.hist(ages, bins=20, edgecolor='red')
plt.title("Histogram of Shark Attack Victim Ages")
plt.xlabel("Age")
plt.ylabel("Number of Victims")
plt.show()
```

```python
shark_data["Type"] = shark_data["Type"].str.title()
shark_data["Type"] = shark_data["Type"].where(
    shark_data["Type"].isin(["Provoked", "Unprovoked"]), "Unknown"
)
(shark_data["Type"] == "Unprovoked").mean()
```

```python
def clean_fatal(x):
    if pd.isna(x):
        return "Unknown"
    x = str(x).upper().strip()  
    if x == "Y":
        return "Y"
    elif x == "N":
        return "N"
    else:
        return "Unknown"

shark_data['Fatal_clean'] = shark_data['Fatal Y/N'].apply(clean_fatal)

print(shark_data['Fatal_clean'].value_counts())
```

```python
fatal_by_type = pd.crosstab(shark_data['Type_clean'], shark_data['Fatal_clean'], normalize='index')
print(fatal_by_type)
```

- The data suggests that the attack is more likely to be fatal when unprovoked/unknown.

**Question 5: All Parts**

```python
college = pd.read_csv("college_completion.csv")
college.head()
college.info()
```

```python
ct = pd.crosstab(college['control'], college['level'])
print(ct)
```

- Private not-for-profit institutions have most 4-years programs, and then public institutions have the most 2-year programs.

```python
plt.figure(figsize=(10,6))
sns.kdeplot(college['grad_100_value'].dropna(), fill=True)
plt.xlabel("Graduation Rate within 100% time (%)")
plt.title("Kernel Density Plot of Graduation Rates")
plt.show()

print(college['grad_100_value'].describe())
```

```python
controls = college['control'].unique()

plt.figure(figsize=(10,6))
for c in controls:
    sns.kdeplot(college.loc[college['control']==c, 'grad_100_value'].dropna(), fill=True, label=c)
plt.xlabel("Graduation Rate within 100% time (%)")
plt.title("Graduation Rate Distribution by Institutional Control")
plt.legend()
plt.show()

for c in controls:
    print(f"\nControl: {c}")
    print(college.loc[college['control']==c, 'grad_100_value'].describe())
```

- Private not-for-profit institutions usually have the highest graduation rates. Public institutions have the lowest graduation rates.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=college, x='aid_value', y='grad_100_value')
plt.xlabel("Average Student Aid ($)")
plt.ylabel("Graduation Rate within 100% time (%)")
plt.title("Graduation Rate vs Average Student Aid")
plt.show()

cov = college[['aid_value','grad_100_value']].cov().iloc[0,1]
corr = college[['aid_value','grad_100_value']].corr().iloc[0,1]
print("Covariance:", cov)
print("Correlation:", corr)
```

- Institutions that provide more aid have higher education rates.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=college, x='aid_value', y='grad_100_value', hue='control')
plt.xlabel("Average Student Aid ($)")
plt.ylabel("Graduation Rate within 100% time (%)")
plt.title("Graduation Rate vs Average Student Aid by Control")
plt.show()

for c in controls:
    subset = college.loc[college['control']==c, ['aid_value','grad_100_value']].dropna()
    print(f"\nControl: {c}")
    print("Covariance:", subset.cov().iloc[0,1])
    print("Correlation:", subset.corr().iloc[0,1])
```

- Private not-for-profit institutions have the strongest positive relationship.

```python
ames = pd.read_csv("ames_prices.csv")
ames.head()
ames.info()
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(ames['price'].dropna(), fill=True)
plt.xlabel("Sale Price ($)")
plt.title("Kernel Density Plot of House Prices")
plt.show()

print(ames['price'].describe())
```

```python
plt.figure(figsize=(10,6))
for bldg in ames['Bldg.Type'].unique():
    sns.kdeplot(ames.loc[ames['Bldg.Type']==bldg, 'price'].dropna(), fill=True, label=bldg)
plt.xlabel("Sale Price ($)")
plt.title("Kernel Density of Price by Building Type")
plt.legend()
plt.show()

price_summary = ames.groupby('Bldg.Type')['price'].describe()
print(price_summary[['count','mean','std','min','25%','50%','75%','max']])
```

- On average, single family homes and end-of-unit townhomes are the most expensive and these also have the highest variance in prices.

```python
def ecdf(data):
    x = np.sort(data)
    y = np.arange(1, len(x)+1)/len(x)
    return x, y

x_ecdf, y_ecdf = ecdf(ames['price'].dropna())
plt.figure(figsize=(10,6))
plt.plot(x_ecdf, y_ecdf, marker='.', linestyle='none')
plt.xlabel("Sale Price ($)")
plt.ylabel("ECDF")
plt.title("ECDF of House Prices")
plt.show()

print("Minimum:", ames['price'].min())
print("25% Quantile:", ames['price'].quantile(0.25))
print("Median:", ames['price'].median())
print("75% Quantile:", ames['price'].quantile(0.75))
print("Maximum:", ames['price'].max())
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(y=ames['price'])
plt.title("Boxplot of House Prices")
plt.show()

plt.figure(figsize=(12,6))
sns.boxplot(x='Bldg.Type', y='price', data=ames)
plt.title("Boxplot of Price by Building Type")
plt.show()
```

- For prices there are a couple of outliers.
- I see that the more expensive houses have outliers and then it slowly decreases with no outliers. 

```python
Q1 = ames['price'].quantile(0.25)
Q3 = ames['price'].quantile(0.75)
IQR = Q3 - Q1

ames['is_outlier'] = (ames['price'] > Q3 + 1.5*IQR)
print(ames['is_outlier'].value_counts())
```

```python
lower = ames['price'].quantile(0.01)
upper = ames['price'].quantile(0.99)

ames['price_winsor'] = ames['price'].clip(lower, upper)

plt.figure(figsize=(10,6))
sns.kdeplot(ames['price_winsor'].dropna(), fill=True)
plt.xlabel("Winsorized Sale Price ($)")
plt.title("Kernel Density Plot of Winsorized Prices")
plt.show()

print(ames['price_winsor'].describe())
```

- The distribution become less skewed.
